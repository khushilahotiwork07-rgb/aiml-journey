```python
import seaborn as sns
'import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
print(type(tips))
print(tips.head())
```

```python
sns.relplot(
    data=tips ,
    x="total_bill" ,
    y="tip",
    hue ="smoker",
    style ="smoker"
)
```

```python
sns.lineplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue = "time")
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
sns.lineplot(
    data=flights,
    x="year",
    y="passengers"
)
```

```python
#bar plot
print(tips.head())
```

```python
sns.barplot(
    data = tips,
    x="day",
    y="tip",
    hue="sex"
)
```

```python
sns.boxplot(
    data= tips,
    x="day",
    y="tip",
    
)
```

```python
#heat maps
print(flights.head())
```

```python
flights_pivot = flights.pivot(index="month" ,columns="year" , values="passengers")
print(flights_pivot)
```

```python
sns.heatmap(
    flights_pivot ,
    cmap="coolwarm"
)
```
