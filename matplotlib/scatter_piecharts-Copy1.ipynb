```python
#scatter graph
import matplotlib.pyplot as plt 
age = [ 20 , 25 , 30 , 40 , 50 , 55 , 60]
bp = [ 80 , 95 , 97 , 120 ,125 , 72 , 135]
plt.scatter(age , bp )
plt.xlabel("age")
plt.ylabel("bp")
plt.title("age vs bp")
```

```python
import matplotlib.pyplot as plt 
age = [ 20 , 25 , 30 , 40 , 50 , 55 , 60]
bp = [ 80 , 95 , 97 , 120 ,125 , 72 , 135]
colors = ["green" if x > 120 else "blue" for x in bp]
plt.scatter(age , bp ,color=colors, s=bp , marker="o" , alpha = 0.3)
plt.xlabel("age")
plt.ylabel("bp")
plt.title("age vs bp")
plt.grid()
```

```python
winter_humidity = [ 20 ,21 ,13 , 16 , 29]
winter_temp = [13, 14 , 15 , 16 ,17]

summer_temp = [ 23, 24 , 25 ,26 ,27]
summer_humidity = [23 , 28 ,19 , 11 , 23]
plt.scatter(winter_humidity , winter_temp , label = "winter_humidity")
plt.scatter(summer_humidity , summer_temp , label = "summer_humidity")
plt.xlabel("humidity")
plt.ylabel("temp")
plt.title("humidity vs temp")
plt.grid()
```

```python
#pie charts
expenses = ["salary" , "electricity bill" , "grocerries" , "education" ,"shopping"]
amount= [90  , 10 , 40 , 50 , 5]
plt.pie(amount , labels=expenses)
```

```python
expenses = ["salary" , "electricity bill" , "grocerries" , "education" ,"shopping"]
amount= [90  , 10 , 40 , 50 , 5]
plt.pie(amount , labels=expenses , autopct = "%1.1f%%" ,
       wedgeprops = {
       "edgecolor" : "black",
           "linewidth" : 0.5,
}
)
plt.show()
```
