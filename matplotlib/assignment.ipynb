```python
import matplotlib.pyplot as plt 
import numpy as np
#compare weekly temprature of 4 cities in a single graph using subplots a
#and loops
days = ["mon" , "tues" , "wed" , "thrus" ,"fri" , "sat" , "sun" ]
cities = [ "tokyo" , "mumbai" ,"newyork" , "london"]
temprature = [ 
    [20 , 24 , 22 , 26 , 27 , 23 , 25],
    [28 , 30 , 36 , 34 , 29 , 30 , 31],
    [18 , 16 , 21 , 20 ,19 , 24 , 19],
    [5 , 6 , 4 , 3 , 9 , 12 , 3]
]
fig, axes = plt.subplots(2 , 2)

city_no = 0
for i in range(2):
    for j in range(2):
        axes [i][j].plot(days , temprature[city_no] ,marker="o" )
        axes[i][j].set_title(cities[city_no])
        city_no = city_no +1
fig.tight_layout
```
