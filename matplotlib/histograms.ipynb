```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
scores = np.random.normal(60, 10 ,100)
plt.hist(scores , color="skyblue" ,  edgecolor="black")
```

```python
marks = [ 78 , 65 , 62 , 33 , 45 ,
          29 , 99 , 77 , 42 , 100, 
          72 , 69 , 92 , 23 , 15 ,
          89 , 40 , 58 , 85 , 66
        ]
bins = [ 10 , 30 , 50 , 70 , 100]
plt.hist( marks , color= "red" , edgecolor="black" , bins=bins)
plt.xlabel("marks")
plt.ylabel("no. of students")
```

```python
#multiple datasets in histogram 
legit_transaction =[49 , 500 , 23.67 , 44 ,26 , 99 , 67,89 , 95 , 56 , 76 , 49 , 21 , 12 ,120 ]
fraud_transaction = [ 200 , 2000 , 450 ,340 ,250 , 600 , 90 , 82 , 700 , 65]
plt.hist(legit_transaction , color="green" , edgecolor ="black",label="legit_transaction" , bins= 10 , alpha =0.5)
plt.hist(fraud_transaction , color="red" , edgecolor = "black" , label = "fraud_transaction" ,bins = 5 , alpha=0.5)
plt.legend()
plt.xlabel("money transaction")
plt.ylabel("no of transaction")
```

```python
#box plots
marks = [4 ,5 ,6 ,7 , 8 ,9 , 10 , 12]
plt.boxplot(marks)
plt.grid()
```

```python
import numpy as np 
group1= np.random.normal(50 ,10 ,100)
group2=np.random.normal(60,20,100)
plt.boxplot([group1, group2] ,label=["group1" , "group2"])
plt.grid()
```

```python
#stack plot 
days = ["mon" , "tues" , "wed" , "thurs" ,"fri" , "sat" , "sun"]
direct= [ 10 , 20 , 30 ,40 , 50 , 60 , 75]
social=[ 20 , 34 , 45 , 30 , 50 , 60 , 66  ]
organic= [15 , 34 , 20 , 12 , 36 , 23 ,72]
plt.stackplot(days , direct , social , organic , labels=["direct" ,"social" , "organic"] )
plt.xlabel("days")
plt.ylabel("no.of customer")
plt.legend()
```

```python
#sub plots
import numpy as np
x= [ 1, 2, 3, 4, 5, 6, 7 ,8 ,9]
y1 = [np.sqrt(i) for i in x]
y2= [ i*2 for i in x]
y3= [i**2 for i in x]
y4 = [i**3 for i in x]

plt.subplot( 2 , 2 ,1)
plt.plot(x , y1)
plt.title("squre root")
plt.subplot( 2 , 2 ,2)
plt.plot(x , y2)
plt.title("double")

plt.subplot( 2 , 2 ,3)
plt.plot(x , y3)
plt.title("squre")

plt.subplot( 2 , 2 ,4)
plt.plot(x , y4)
plt.title("cube")

```

```python
#mordern matplotlib
fig , ax= plt.subplots(2,2)
ax[0][0] .plot(x,y1)
ax[0][0].set_title("square root")

ax[0][1].plot(x,y2)
ax[0][1].set_title("double")
```
