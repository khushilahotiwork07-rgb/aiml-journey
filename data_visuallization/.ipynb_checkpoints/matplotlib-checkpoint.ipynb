```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2 , 3 , 4 , 5]
y = [2 , 4 , 6 , 8 , 10]
plt.plot(x,y)
```

```python
#real data
x = (2008 , 2009 , 2010 , 2011 )
y = (100 , 105 , 88 , 99 )

a = (2008 , 2009, 2010 , 2011)
b = (110 , 120 ,97 ,105)
plt.plot(a , b  , color = "red" , marker = "<" ,label = "non oscar movies")
plt.plot(x , y , "o--c", label = "oscar movies")
plt.title("oscare movies vs non oscars movies revenue in years")
plt.xlabel("years")
plt.ylabel("revenue in millions")
plt.legend()
```

```python
x = (2008 , 2009 , 2010 , 2011 )
y = (100 , 105 , 88 , 99 )

a = (2008 , 2009, 2010 , 2011)
b = (110 , 120 ,97 ,105)
plt.plot(a , b  ,label = "non oscar movies")
plt.plot(x , y ,label = "oscar movies")
plt.title("oscare movies vs non oscars movies revenue in years")
plt.xlabel("years")
plt.ylabel("revenue in millions")
plt.legend()
plt.grid()
```
