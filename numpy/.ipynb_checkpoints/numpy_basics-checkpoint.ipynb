```python
#ones array
import numpy as np
arr = np.ones((3,3))
print(arr)
```

```python
#full array
arr = np.full((2, 3 ),20)
print(arr)
```

```python
#identity matrix 
arr = np.eye(3)
print(arr)
```

```python
#range function
arr = np.arange(1 , 10)
print(arr)
```

```python
#even no
arr = np.arange(0 , 10 , 2)
print(arr)
#skip every 3rd no
arr1 = np.arange(1 , 14 , 2)
print(arr1)
#linespace
arr3 = np.linspace(1 , 100 , 4)
print(arr3)
```

```python
#reshaping of array
arr = np.array([[1,2,3] , [4,5,6]])
print(arr , len(arr) , arr.shape)

reshaped = arr.reshape((3 , 2))
print(reshaped , reshaped.shape)
#flattend
flattend = arr.flatten()
print(flattend)
```

```python
#indexing
arr = np.array([1 , 2 , 3 ,4 , 5])
print(arr[2])

#fancy indexing
idx = [0 , 1 , 3]
print(arr[idx])

#boolean indexing 
print(arr[arr>3])

#slicing
print(arr[1:3])
print(arr.dtype)
```
