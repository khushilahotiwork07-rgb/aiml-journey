```python
import numpy as np
import pandas as pd
df = pd.read_csv("IRIS.csv")
df.head(10)
df.shape
df.info()
df["petal_length"].mean()
df["petal_length"].max()
```

```python
#Select only those rows where:
#Select only those rows where:
#• petal_length > 4.5
#• species = "Iris-virginica"
df[(df["petal_length"] > 4.5) & (df["species"] == ("Iris-virginica"))]

```

```python
#Group by species and compute:
#• Average sepal_length
# Maximum petal_width
#• Standard deviation of sepal_width
df.groupby("species").agg({
    "sepal_length" : "mean", 
    "petal_width" : "max",
    "sepal_width" : "std"
})
    
```

```python
#Create a new column: "petal_ratio" = petal_length / petal_width
#• For each species, find the average petal_ratio.

df["petal_ratio"] = df["petal_length"] / df["petal_width"]
df
```
