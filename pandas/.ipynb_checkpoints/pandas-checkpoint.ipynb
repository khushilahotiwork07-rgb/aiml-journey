```python
#dataframe 
#dictionary of lists
info = { 
    "name" : ["khushi" , "rahul" , "abhishek"],
    "class" : [12 , 11 , 11],
    "gpa" : [2.3 , 5.5 , 9.8]
}
df = pd.DataFrame(info)
print(df)
print(df.index)
```
